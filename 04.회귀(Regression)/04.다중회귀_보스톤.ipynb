# 보스톤 주택가격

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
df['PRICE'] = boston.target
df.head()
```

* 산점도와 선형회귀선

```python
sns.regplot(x='CRIM', y='PRICE', data=df)
plt.show()
```

```python
_, axs = plt.subplots(nrows=3, ncols=4, figsize=(16,12))
for i, feature in enumerate(df.columns[1:-1]):
    row, col = i//4, i%4
    sns.regplot(x=feature, y='PRICE', data=df, ax=axs[row][col])
```

- feature별 R squared valuem

```python
from sklearn.linear_model import LinearRegression
```

```python
# CRIM
X = df.CRIM.values.reshape(-1,1)
y = boston.target
lr = LinearRegression()
lr.fit(X,y)
lr.score(X,y)
```

```python
y = boston.target
for feature in df.columns[:-1]:
    X = df[feature].values.reshape(-1,1)
    lr = LinearRegression()
    lr.fit(X,y)
    score = lr.score(X, y)
    print(f'{feature}:\t{score:.4f}')
```

```python
# 오차가 적을수록 회귀선 근처에 모여있어. RM, LSTAT 등등. 우린 CRIM에 대해서도 봤어 .
```

* 다중선형회귀

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    boston.data, boston.target, test_size=0.1, random_state=2021
)
```

```python
lr=LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 다중 선형회귀에서는 R squared value가 무의미
lr.score(X_test, y_test)
```

```python
# Weight
lr.coef_
```

```python
# Bias
lr.intercept_
```

- 회귀식
    <pre>
        PRICE = -0.11 * CRIM + 0.053 * ZN + ... + -0.55 * LSTAT + 35.5
    <pre>

* 테스트테이터

```python
X_test[0], y_test[0]
```

```python
# 예측값 구하기. 회귀식은 위에 구했어 ? 
np.dot(X_test[0], lr.coef_) + lr.intercept_
```

```python
# predict로 예측한 값
lr.predict(X_test[:1])

# x data로 2차원 데이터를 원해. 위처럼 하거나 리쉐입 하거나
```

```python
# 10개의 데이터에 대한 예측값과 실제값
for i in range(10):
    pred1 = np.dot(X_test[i], lr.coef_) + lr.intercept_
    pred2 = lr.predict(X_test[i].reshape(1,-1))
    # pred2 = lr.predict([X_test[0]])
    # pred2 = lr.predict(X_test[i:i+1]
    print(f'실제값: {y_test[i]}, 예측값1 :{pred1:.2f}, 예측값2 : {pred2[0]:.2f}')
```
