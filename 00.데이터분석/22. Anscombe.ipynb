Anscombe's quartet  앤스컴 콰르텟

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
ans = sns.load_dataset('anscombe')
ans.shape
```

```python
# ans.head()
ans.info()
```

```python
ans
```

데이터셋 1 에대해서 그래프를 그려봄

```python
ds = ans[ans.dataset == 'I']
ds
```

- 회귀식이 y = 0.5 x + 3 일것

```python
plt.scatter(ds.x, ds.y)
plt.plot([4,14], [5,10], 'r-', lw=2)
plt.title('Dataset I')
plt.show()
```

```python
# 4개의 그래프 그리기 # 하나의 그림에 그리기

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12,10) )
for i, roman in enumerate(['I','II', 'III','IV']):
    ax = axs[i//2, i%2]
    
    ds = ans[ans.dataset == roman]
    ax.scatter(ds.x, ds.y)
    ax.plot([3,20], [4.5,13], 'r-', lw=2)
    ax.set_ylim(2,14)
    ax.set_title(f'Dataset{roman}')
```

```python
# 모양 좀더 이쁘게 # sharex 와 y사용. 슈퍼타이틀 타이트 레이아웃.

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12,10), sharex=True, sharey=True )
for i, roman in enumerate(['I','II', 'III','IV']):
    ax = axs[i//2, i%2]
    
    ds = ans[ans.dataset == roman]
    ax.scatter(ds.x, ds.y)
    ax.plot([3,20], [4.5,13], 'r-', lw=2)
    ax.set_ylim(2,14)
    ax.set_title(f'Dataset{roman}')

plt.suptitle("Ansconbe's Quartet", fontsize=20)
plt.tight_layout()
plt.show()
```
