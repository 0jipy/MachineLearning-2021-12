### 1번 데이터프레임 처리과정 (구별 인구데이터)

```python
# 서울시 구별 주민등록정보 데이터를 s로 이름
s = pd.read_csv('서울시주민등록인구(구별)통계.txt', skiprows=2, sep='\t')
s.head()
```

```python
# 1번데이터 전처리    #컬럼별 결측값 없음을 확인할 수 있다.
# s.isna().sum()      #컬럼별 결측값.
s.isna().sum().sum()  #전체 결측값. 확인차.
# s.info()  #인구 값들이 문자열이라 int 변환및   1000단위의  컴마 , 제거할것.
```

```python
s.columns
```

```python
s = s[['자치구', '계', '계.2', '65세이상고령자']]   # 컬럼추출
# s.head(3)
```

```python
s.columns = ['구별', '인구수', '외국인', '고령자']    # 컬럼이름 변경
# s.head(3)
```

```python
# , 없애기 및 정수변환 / 주의할점은 s.columns[1:] 의미.
for column in s.columns[1:]:
    s[column] = s[column].apply(lambda x : int(x.replace(',','')))
# s.info()  #정수로 잘 변경됨 확인
```

```python
# 외국인비율(%) 및 고령자비율(%) 추가
# 값 자릿수 고려. 소숫점이하 2자리까지. 
s['외국인비율(%)'] = np.round((s.외국인 / s.인구수) * 100, 2)
s['고령자비율(%)'] = np.round((s.고령자 / s.인구수) * 100, 2)
# 이거 한번에 되나? #둘이 결합하는 방법
# s.head(3)
```

```python
s.drop(columns= ['외국인','고령자'],inplace=True)
s.head(3)
```

```python
# c.isnull()
c.isna().sum().sum()
# 결측치 없다. info는 생략. 
```

```python
for column in c.columns[1:]:
    c[column] = c[column].apply(lambda x : int(x.replace(',','')))
c.head()
```

```python
c.columns
```

```python
# 최근증가율 컬럼 추가. 다만 %라고 생각하기엔 애매해서 '최근증가율'로 마무리 
c['최근증가율'] = ((( c['2018년'] + c['2019년'] + c['2020년'] ) / (
                                 c['2012년 이전'] + c['2013년'] + c['2014년'] + c['2015년'] + c['2016년'] +c['2017년'] ))*100).round(2)
# c.head()
```

```python
c = c[['구분', '총계', '최근증가율']]
c.head()
```

```python
# df.rename(columns={'Before':'After'}) # 열 선택하여 바꿀때 
# df.columns = ['col', 'col', 'col']    # 전체컬럼이름 바꿀때는 요게 편하다
c.columns = ['구별', 'CCTV댓수', '최근증가율']
c.head(3)
```

```python
# 구이름 공백지우기 # 중구 데이터 하나만 '중 구'로 띄어쓰기 되어있어서 처리 필요
c['구별'] = c.구별.apply(lambda x: x.replace(' ',""))
c.head()
```

### 3. 공통 컬럼 및 인덱싱 확인 merge 이용해 붙이기
 - cctv 데이터인 c에 서울시 인구데이터 s 를 붙일것 : 주요데이터가 CCTV댓수

```python
cctv = pd.merge(c, s)
```

```python
# 교재 63p처럼 '구별'컬럼 인덱스로 만들기
cctv.set_index('구별')
cctv
```

```python
import numpy as np
import pandas as pd
```

```python
# cctv data 
cctv = pd.read_csv('서울시CCTV설치운영현황(자치구)_년도별_210731기준.csv', skiprows=1, encoding='euc-kr'
)
cctv.head()
```

```python
# 결측치 확인 결과는 0
```

```python
# 1000단위 구분기호 없애기
# cctv.columns
for column  in cctv.columns[1:]:
    cctv[column] = cctv[column].apply(lambda x : int(x.replace(',','')))
cctv.head()
```

```python
# 최근 증가율 
cctv['최근3개년'] = cctv['2018년'] + cctv['2019년'] + ['2020년']
cctv['2017년이전'] = cctv.총계 - cctv['최근3개년'] - cctv['2021년']
cctv['최근증가율'] = (cctv['최근3개년'] / cctv['2017년이전'] * 100).round(2)
cctv.head()
```

```python
cctv.head()
```

```python
cctv['최근증가율2'] = (cctv.iloc[:,9:12].sum(axis=1)/cctv.iloc[:,2:9].sum(axis=1)*100).round(2)
cctv.head()
```

```python
# 셀렉션 필요한 컬럼 추출
cctv = cctv[['구분', '총계', '최근증가율2']]
# cctv
```

```python
# 컬럼병 변경 구분 을 구별 총계를 ㅊㅊㅅㅍ 댓수. 최근증가율2 를 최근증가율로 
cctv.rename(columns={'구분':'구별', '총계' : 'CCTV댓수', '최근증가율2' : '최근증가율'}, inplace=True)
cctv.head()
```

```python
# 첫번째 행(계) 삭제
cctv.drop([0], inplace=True)
cctv.head()
```

```python
# 구이름에서 공백 지우기 #오오 str?이뭐닝.
cctv.구별 = cctv.구별.str.replace(' ','')
cctv.head()
```

```python
# cctv 댓수가 많은 자치구 5 # 데이터 프레임에서 소팅

cctv.sort_values('CCTV댓수', ascending=False).head()
```

# 2. 서울시 인구 데이터

```python
pop = pd.read_csv('서울시주민등록인구(구별)통계.txt', sep='\t', skiprows=2 )
# pop
```

```python
pop.isna().sum().sum()
```

```python
# 필요한 컬럼 추출 및 이름 변경
pop = pop[['자치구', '계','계.2', '65세이상고령자']]
pop.columns = ['구별', '인구수', '외국인', '고령자']
pop.head(3)
```

```python
# 천단위 구분기호 없애기
for column in pop.columns[1:]:
    pop[column] = pop[column].str.replace(',','').astype(int)
pop.head(3)
```

```python
# 첫번째 행 지우기 
pop.drop([0], inplace=True)
```

```python
# 외국인비율, 고령자비율 컬럼 만들기
pop['외국인비율'] = (pop.외국인 / pop.인구수 * 100).round(2)
pop['고령자비율'] = (pop.고령자 / pop.인구수 * 100).round(2)
```

```python
pop
```
